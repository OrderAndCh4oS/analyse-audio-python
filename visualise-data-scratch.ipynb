## Heatmaps

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.DataFrame([
    [0.75, 0.25, 0],
    [0, 1, 0],
    [0, 0, 1],
])
df.columns = ['one', 'two', 'three']
df_corr = df.corr()

f, ax = plt.subplots(figsize=(10, 10))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(df_corr, annot=True, cmap=cmap)
```
